```python
'''A manufacturing company is concerned about the production times of two different machines.
Compare the distribution of the production times for each machine using boxplots.
Identify any outlier access the central tendency in the production process'''

import matplotlib.pyplot as plt

# Sample production times for Machine 1 and Machine 2
machine_1_times = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]
machine_2_times = [8, 10, 11, 13, 14, 15, 16, 17, 18, 20]

# Calculate central tendency (mean) for each machine
machine_1_mean = sum(machine_1_times) / len(machine_1_times)
machine_2_mean = sum(machine_2_times) / len(machine_2_times)

# Create boxplots
plt.boxplot([machine_1_times, machine_2_times], labels=['Machine 1', 'Machine 2'], notch=True)

# Identify potential outliers (outside 1.5 IQR from the quartiles)
q1_machine_1 = sorted(machine_1_times)[int(len(machine_1_times) * 0.25)]
q3_machine_1 = sorted(machine_1_times)[int(len(machine_1_times) * 0.75)]
iqr_machine_1 = q3_machine_1 - q1_machine_1
lower_bound_machine_1 = q1_machine_1 - (1.5 * iqr_machine_1)
upper_bound_machine_1 = q3_machine_1 + (1.5 * iqr_machine_1)

q1_machine_2 = sorted(machine_2_times)[int(len(machine_2_times) * 0.25)]
q3_machine_2 = sorted(machine_2_times)[int(len(machine_2_times) * 0.75)]
iqr_machine_2 = q3_machine_2 - q1_machine_2
lower_bound_machine_2 = q1_machine_2 - (1.5 * iqr_machine_2)
upper_bound_machine_2 = q3_machine_2 + (1.5 * iqr_machine_2)

outliers_machine_1 = [point for point in machine_1_times if point < lower_bound_machine_1 or point > upper_bound_machine_1]
outliers_machine_2 = [point for point in machine_2_times if point < lower_bound_machine_2 or point > upper_bound_machine_2]

# Add labels and title
plt.xlabel('Machine')
plt.ylabel('Production Time (Units)')
plt.title('Comparison of Production Times for Two Machines')

# Print central tendency and potential outliers
print(f"Machine 1 Mean: {machine_1_mean:.2f}")
print(f"Machine 2 Mean: {machine_2_mean:.2f}")
print(f"Machine 1 Potential Outliers: {outliers_machine_1}")
print(f"Machine 2 Potential Outliers: {outliers_machine_2}")

plt.show()
```

```python
'''A retail store wants to analyze the sales performance of its products across multiple
categories. generate a barplot to compare the total sales for each product category. highlight
the top performance categories and provide insight to enhance the store product offerings.'''

import matplotlib.pyplot as plt

# Product categories and their total sales
categories = ["Clothing", "Electronics", "Furniture & Homeware", "Beauty & Personal Care", "Grocery & Food", "Toys & Games", "Sports & Fitness", "Garden & Outdoors"]
sales = [15000, 22000, 18000, 12000, 20000, 8000, 10000, 5000]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(categories, sales, color=['skyblue', 'orange', 'green', 'purple', 'red', 'yellow', 'blue', 'brown'])
plt.xlabel("Product Category")
plt.ylabel("Total Sales ($)")
plt.title("Sales Performance by Category")
plt.xticks(rotation=45, ha='right')  # Rotate category names for better readability

# Highlight top 2 performing categories
plt.bar(categories[:2], sales[:2], color='gold', label='Top Performers')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
'''A car dealership is interested in understanding the relationship between the mileage
and resale value of used cars in the inventory. collect data on the mileage and resale value 
of various car models and  create a scatterplot. analyze the plot to identify any correlation 
patterns and provide insights into how mileage influences resale value, helping the dealership
make informed pricing decision.
'''

import matplotlib.pyplot as plt

# Sample data for mileage and resale value of used cars
mileage = [30000, 50000, 70000, 90000, 110000, 130000, 150000]
resale_value = [18000, 15000, 12000, 9000, 7000, 5000, 3000]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(mileage, resale_value, color='blue', marker='o', edgecolors='black')
plt.xlabel('Mileage (in thousands)')
plt.ylabel('Resale Value ($)')
plt.title('Relationship Between Mileage and Resale Value of Used Cars')

# Analyze the plot: A negative correlation is observed, indicating that higher mileage generally leads to lower resale value.

print("The scatter plot suggests a negative correlation between mileage and resale value. As the mileage increases, the resale value tends to decrease.")

plt.grid(True)
plt.show()
```
