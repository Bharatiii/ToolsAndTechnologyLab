```python
''' Create two NumPy arrays arr1 and arr2 of size 5x5 with random integer
values. Perform element-wise addition, subtraction, multiplication, and division between
arr1 and arr2.
'''
import numpy as np

# Create two NumPy arrays with random integer values
arr1 = np.random.randint(1, 10, size=(5, 5))  # Random integers between 1 and 10
arr2 = np.random.randint(1, 10, size=(5, 5))  # Random integers between 1 and 10

# Perform element-wise addition
addition = arr1 + arr2

# Perform element-wise subtraction
subtraction = arr1 - arr2

# Perform element-wise multiplication
multiplication = arr1 * arr2

# Perform element-wise division (watch out for division by zero)
# Here, we'll replace division by zero with zero itself to avoid runtime warnings
with np.errstate(divide='ignore', invalid='ignore'):
    division = np.where(arr2 != 0, arr1 / arr2, 0)

# Displaying the arrays and the results
print("Array 1:")
print(arr1)
print("\nArray 2:")
print(arr2)

print("\nElement-wise addition:")
print(addition)

print("\nElement-wise subtraction:")
print(subtraction)

print("\nElement-wise multiplication:")
print(multiplication)

print("\nElement-wise division:")
print(division)
```

```python
''' Generate a sequence of numbers from 0 to 10 and their corresponding square
values. Create a line plot to visualize the relationship between the numbers and their
squares.
'''
import numpy as np
import matplotlib.pyplot as plt

# Generate sequence of numbers from 0 to 10
numbers = np.arange(0, 11)

# Compute square values
squares = numbers ** 2

# Create line plot
plt.plot(numbers, squares, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Numbers')
plt.ylabel('Square Values')
plt.title('Relationship Between Numbers and Their Squares')

# Display the plot
plt.grid(True)  # Add grid
plt.show()
```

```python
''' Generate two arrays of random numbers, x and y, each of size 100. Create a
scatter plot to visualize the relationship between x and y.'''

import numpy as np
import matplotlib.pyplot as plt

# Generate arrays of random numbers
x = np.random.rand(100)
y = np.random.rand(100)

# Create scatter plot
plt.scatter(x, y, color='blue', alpha=0.7)  # alpha parameter controls transparency

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of X vs Y')

# Display the plot
plt.grid(True)  # Add grid
plt.show()
```

```python
'''Problem: Generate an array of random numbers from a normal distribution. Create a
histogram using Seaborn to visualize the distribution of these numbers.
'''

import seaborn as sns
import numpy as np

# Define the parameters of the normal distribution
mean = 50  # Adjust this for your desired mean
std_dev = 10  # Adjust this for your desired standard deviation
sample_size = 1000  # Number of random samples

# Generate random numbers from the normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Create the histogram using Seaborn
sns.distplot(data)

# Customize the plot (optional)
# You can add title, labels, etc. using Matplotlib functions
plt.title("Histogram of Random Numbers from Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
'''Problem: Generate a correlation matrix from a Pandas DataFrame. Create a heatmap
using Seaborn to visualize the correlation between different numerical variables in the
DataFrame'''
import seaborn as sns
import pandas as pd

# Load your dataset into a pandas DataFrame (replace 'data.csv' with your filename)
df = pd.read_csv("C:/Users/KIIT/Desktop/kaggle/dataset/iris.data.csv")

# Select only numerical columns (optional)
numerical_cols = df.select_dtypes(include=[np.number])  # Select numerical columns

# Calculate correlation matrix
corr_matrix = numerical_cols.corr()

# Create heatmap using Seaborn
sns.heatmap(corr_matrix, annot=True)  # Set annot=True to display correlation values

# Customize the plot (optional)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```
