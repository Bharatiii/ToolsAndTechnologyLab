```python
'''WAP in python to Plot a simple line graph representing the values [1,4,9,16,25].
'''

import matplotlib.pyplot as plt

# Define the data lists
x = [1, 2, 3, 4, 5]  # Assuming x-axis values are consecutive integers (1 to 5)
y = [1, 4, 9, 16, 25]

# Create the line plot
plt.plot(x, y)

# Add labels to the axes
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a title to the graph
plt.title("Line Graph")

# Display the plot
plt.show()
```

```python
'''WAP in python to Create a scatter plot with random x and y values.'''

import matplotlib.pyplot as plt
import numpy as np

# Define the number of data points
num_points = 100

# Generate random x and y values between 0 and 1
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Create the scatter plot
plt.scatter(x, y)

# Add labels to the axes
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a title to the graph
plt.title("Scatter Plot with Random Data")

# Display the plot
plt.show()
```

```python
'''WAP in python Use seaborn to create to create a barplot for categories A,B and C 
with corresponding [3,7,5].'''

import seaborn as sns
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [3, 7, 5]


sns.barplot(x=categories, y=values, color='skyblue')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot with Seaborn')


plt.show()
```

```python
'''WAP in python to Generate a dataset of random values and create a histogram using seaborn.'''
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Define the number of data points
num_points = 1000

# Generate random data between 0 and 100
data = np.random.rand(num_points) * 100

# Create the histogram
sns.distplot(data)

# Add a title and labels to the plot
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
'''WAP in python to Create a boxplot using seaborn for a dataset with multiple categories.
'''
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample data with categories and values
data = {
    'category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'value': [2, 5, 7, 1, 4, 8, 3, 6, 9]
}

# Create a DataFrame from the data dictionary
df = pd.DataFrame(data)

# Create the boxplot
sns.boxplot(
    x = "category",
    y = "value",
    showmeans=True,  # Optional: to show means as points on the plot
    data=df
)

# Add a title and rotate x-axis labels for readability if needed
plt.title("Boxplot of Values by Category")
plt.xticks(rotation=45)  # Uncomment to rotate x-axis labels

# Display the plot
plt.show()
```

```python
'''.WAP in python to reshape a 1D array with 12 elements into the 2Darray with shape(3,4).
'''


import numpy as np

# Create a 1D array with 12 elements
array = np.arange(12)  # Creates an array from 0 to 11

# Reshape the array into a 2D array with shape (3, 4)
reshaped_array = array.reshape(3, 4)

# Print the original and reshaped arrays
print("Original array:\n", array)
print("\nReshaped array:\n", reshaped_array)
```

```python
'''Perform reshaping ((1* 9) and (9*1)) and and transposing on the above two matrices'''

import numpy as np

# Create a 1x9 array
matrix_1x9 = np.arange(1, 10)  # Elements from 1 to 9

# Print the original 1x9 matrix
print("Original 1x9 matrix:\n", matrix_1x9)

# Reshape operations
# Reshape to 9x1 (no actual change in data as it's a 1D array)
reshaped_matrix_9x1 = matrix_1x9.reshape(9, 1)

# Transpose operations
# Transpose the 1x9 matrix (remains 1x9)
transposed_matrix_1x9 = matrix_1x9.T

# Print the reshaped and transposed matrices
print("\nReshaped matrix (9x1):\n", reshaped_matrix_9x1)
print("\nTransposed matrix (1x9):\n", transposed_matrix_1x9)
```
